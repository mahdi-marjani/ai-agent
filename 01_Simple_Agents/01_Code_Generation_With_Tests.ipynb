```python
import subprocess
import tempfile
```

```python
with open("./01_test.py", "r") as f:
    TEST_CODE = f.read()
```

```python
TASK = """
You are a coding agent.

Task:
Write a Python function factorial(n) that returns n factorial.

Rules:
- Return ONLY valid Python code
- Do NOT explain anything
- The code must define factorial(n)
"""
```

```python
def ask_llm(previous_error=None, previous_code=None):
    prompt = TASK
    
    if previous_error:
        prompt += f"\n\nPrevious code:\n{previous_code}\n\nError:\n{previous_error}\n\nUpdate the code to fix the error."

    print("-" * 50)
    print("PROMPT SENT TO LLM:")
    print("-" * 50)
    print(prompt)
    print("-" * 50)

    input("Press Enter after updating response.txt ...")

    with open("./01_response.txt", "r") as f:
        response = f.read().strip()

    # print("-" * 50)
    print("RESPONSE FROM LLM:")
    print("-" * 50)
    print(response)
    print("-" * 50)

    return response
```

```python
def run_code(code):
    with tempfile.NamedTemporaryFile("w", suffix=".py", delete=False) as f:
        f.write(code)
        f.write("\n\n")
        f.write(TEST_CODE)
        filename = f.name

    result = subprocess.run(["python", filename], capture_output=True, text=True)

    return result.returncode == 0, result.stderr
```

```python
error = None
code = None
iteration = 0

while True:
    iteration += 1
    print(f"\n{'='*20} Iteration {iteration} {'='*20}\n")

    code = ask_llm(error, code)
    
    success, error = run_code(code)

    if success:
        print("\n✅ Tests passed! Agent finished successfully.")
        break
    else:
        print("\n❌ Tests failed. Retrying...")
```
